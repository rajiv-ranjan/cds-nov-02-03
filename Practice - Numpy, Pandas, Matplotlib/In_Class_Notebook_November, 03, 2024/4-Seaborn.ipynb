#### Errorbar in seaborn plot
- Error Bars can be applied to graphs such as 
    - Scatterplots, Dot Plots, Bar Charts or Line Graphs
- Error Bars provide an additional layer of detail on the presented data.
- It help to indicate estimated error or uncertainty to give a general sense of how precise a measurement is.

#### How the Errorbar is displayed ?
- Error Bars are displayed through the use of markers drawn over the original graph and its data points.
- Typically, Error bars are used to display either the 
    - standard deviation, 
    - standard error, 
    - confidence intervals or 
    - minimum and maximum values in a ranged dataset.
- The **length of an Error Bar helps reveal the uncertainty of a data point**

- By default, the barplot() function in seaborn draws error bars in the plot with 95% confidence interval. 
- A confidence interval, in statistics, refers to the probability that a population parameter will fall between two set values.
- You can remove error bars by passing errorbar=None argument.

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Bar plot of titanic dataset
# Display average Fare by Class and Gender on the Titanic

titanic = sns.load_dataset("titanic")

# customize the bar plot

#sns.barplot(x='class', y='fare', data=titanic, hue='sex')

#sns.barplot(x='class', y='fare', data=titanic, hue='sex', errorbar='sd')

sns.barplot(x='class', y='fare', data=titanic, hue='sex', errorbar=None)

plt.show()
```

#### Kernel density plot - with fill parameter
- Density plots are also known as kernel density plots
- They are a type of data visualization that display the distribution of a continuous variable. 
- They are similar to histograms, 
    - but instead of representing the data as bars, density plots use a smooth curve to estimate the density of the data. 
- In Seaborn, density plots can be created using the kdeplot() function.

```python
# Create a density plot of the "total_bill" column from the "tips" dataset

# Load the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

sns.kdeplot(data=tips, x='total_bill', hue='time', fill=True)

plt.show()
```
